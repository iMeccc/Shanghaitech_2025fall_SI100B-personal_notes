```python
import sys

def input_matrix(rows, cols, name):
    """è¾“å…¥çŸ©é˜µæ•°æ®ï¼ˆå¸¦æ•´æ•°æ ¡éªŒï¼‰"""
    matrix = [[0 for _ in range(cols)] for _ in range(rows)]
    for i in range(rows):
        for j in range(cols):
            while True:
                try:
                    val = int(input(f"{name}[{i+1}][{j+1}]="))
                    matrix[i][j] = val
                    break
                except ValueError:
                    print("è¯·è¾“å…¥æ•´æ•°ï¼", end=" ")
    return matrix

def matrix_multiply(A, B):
    """çŸ©é˜µä¹˜æ³•ï¼šAÃ—Bï¼Œè¿”å›ç»“æœçŸ©é˜µ"""
    row_A, col_A = len(A), len(A[0]) if A else 0
    row_B, col_B = len(B), len(B[0]) if B else 0

    # åˆå§‹åŒ–ç»“æœçŸ©é˜µå¹¶è®¡ç®—æ¯ä¸ªå…ƒç´ 
    result = []
    for i in range(row_A):
        current_row = []  # å­˜å‚¨ç»“æœçŸ©é˜µçš„ç¬¬iè¡Œ
        for j in range(col_B):
            # æå–Bçš„ç¬¬jåˆ—ï¼š[B[0][j], B[1][j], ..., B[row_B-1][j]]
            col_B_j = [B[k][j] for k in range(row_B)]
            # è®¡ç®—Açš„ç¬¬iè¡Œä¸Bçš„ç¬¬jåˆ—çš„ç‚¹ç§¯ï¼ˆæ ¸å¿ƒæ­¥éª¤ï¼‰
            dot_product = 0
            for a, b in zip(A[i], col_B_j):
                dot_product += a * b
            current_row.append(dot_product)
        result.append(current_row)
    return result

def print_matrix(matrix, name="ç»“æœçŸ©é˜µ"):
    """æ ¼å¼åŒ–æ‰“å°çŸ©é˜µï¼Œç¡®ä¿å¯¹é½"""
    print(f"\n{name} =")
    for row in matrix:
        print(" ".join(f"{num:6}" for num in row))  # æ¯ä¸ªå…ƒç´ å 6ä½å®½åº¦

if __name__ == "__main__":
    # è¾“å…¥çŸ©é˜µç»´åº¦ï¼ˆå¸¦æ ¡éªŒï¼‰
    try:
        row_A = int(input("çŸ©é˜µAè¡Œæ•°ï¼š"))
        col_A = int(input("çŸ©é˜µAåˆ—æ•°ï¼š"))
        row_B = int(input("çŸ©é˜µBè¡Œæ•°ï¼š"))
        col_B = int(input("çŸ©é˜µBåˆ—æ•°ï¼š"))
    except ValueError:
        print("ç»´åº¦å¿…é¡»æ˜¯æ•´æ•°ï¼")
        sys.exit(1)

    # æ£€æŸ¥ä¹˜æ³•å¯è¡Œæ€§
    if col_A != row_B:
        print(f"æ— æ³•ç›¸ä¹˜ï¼šAåˆ—æ•°({col_A})â‰ Bè¡Œæ•°({row_B})")
        sys.exit(1)

    # è¾“å…¥çŸ©é˜µæ•°æ®
    matrix_A = input_matrix(row_A, col_A, "A")
    matrix_B = input_matrix(row_B, col_B, "B")

    # è®¡ç®—å¹¶æ‰“å°ç»“æœ
    result = matrix_multiply(matrix_A, matrix_B)
    print_matrix(result, "A Ã— B")
    
```

```python
# get the input
order = int(input("the order of the determinant is:"))

def det_init(order):
    det = [[0 for _ in range(order)]for _ in range(order)]
    return det

# do the calculation

def two_order_cal(det):
    result = det[0][0]*det[1][1]-det[0][1]*det[1][0]
    return result

def det_cal(det,order):
    final_result = 0
    if order > 2:    
        new_order = order-1
        result = [0 for _ in range(order)]
        for i in range(order):
            new_det = det_init(new_order)
            for p in range(new_order):
                for q in range(new_order):
                    if q < i:
                        new_det[p][q] = det[p+1][q]
                    elif q >= i:
                        new_det[p][q] = det[p+1][q+1]

        # code above has realised the input of the new determinant
        # code below will do the calculation
            result[i] = det[0][i]*det_cal(new_det,new_order)
            if i%2 == 0:
                result[i] = -result[i]
            final_result += result[i]

    elif order == 2:
        final_result = two_order_cal(det)

    return final_result

determinant = det_init(order)
for i in range(order):
        for j in range(order):
            determinant[i][j] = int(input(f"det[{i+1}][{j+1}]="))
result = det_cal(determinant,order)
print(f"The determinant equals to {result}")
```

- èªæ˜çš„è±†åŒ…åˆ©ç”¨é«˜æ–¯æ¶ˆå…ƒæ³•å†™å‡ºäº†æ–°çš„ç®—æ³•â€¼ï¸å¤æ‚åº¦ä»…ä¸ºO(n^3)â€¼ï¸
> é«˜æ–¯æ¶ˆå…ƒæ³•åŸç†ï¼šå…ˆè½¬æ¢ä¸ºä¸Šä¸‰è§’çŸ©é˜µï¼Œè¡Œåˆ—å¼å€¼ = ä¸»å¯¹è§’çº¿å…ƒç´ çš„ä¹˜ç§¯ Ã— è¡Œäº¤æ¢å¯¼è‡´çš„ç¬¦å·ï¼ˆæ­£/è´Ÿï¼‰  
å…¶ä¸­æ¯è¿›è¡Œä¸€æ¬¡è¡Œäº¤æ¢ï¼Œæ­£è´Ÿå·å¯¹è°ƒä¸€æ¬¡ã€‚

```python
def init_matrix(order):
    """åˆå§‹åŒ–orderÃ—orderçš„é›¶çŸ©é˜µ"""
    return [[0 for _ in range(order)] for _ in range(order)]


def determinant_calculator(matrix, order):
    """
    ä½¿ç”¨é«˜æ–¯æ¶ˆå…ƒæ³•è®¡ç®—è¡Œåˆ—å¼å€¼
    åŸç†ï¼šå°†çŸ©é˜µåŒ–ä¸ºä¸Šä¸‰è§’çŸ©é˜µåï¼Œå¯¹è§’çº¿å…ƒç´ ä¹˜ç§¯å³ä¸ºè¡Œåˆ—å¼å€¼ï¼ˆéœ€è€ƒè™‘è¡Œäº¤æ¢æ¬¡æ•°å¯¹ç¬¦å·çš„å½±å“ï¼‰
    """
    # æ·±æ‹·è´çŸ©é˜µé¿å…ä¿®æ”¹åŸçŸ©é˜µ
    mat = [row.copy() for row in matrix]
    sign = 1  # è®°å½•è¡Œäº¤æ¢æ¬¡æ•°çš„ç¬¦å·ï¼ˆå¶æ•°æ¬¡ä¸º1ï¼Œå¥‡æ•°æ¬¡ä¸º-1ï¼‰

    for col in range(order):  # éå†æ¯ä¸€åˆ—ï¼ˆä¸»å…ƒåˆ—ï¼‰
        # æ­¥éª¤1ï¼šæ‰¾åˆ°å½“å‰åˆ—ä¸­ä¸‹æ–¹ï¼ˆåŒ…æ‹¬å½“å‰è¡Œï¼‰ç»å¯¹å€¼æœ€å¤§çš„è¡Œä½œä¸ºä¸»è¡Œï¼ˆå‡å°‘ç²¾åº¦è¯¯å·®ï¼‰
        pivot_row = col
        for row in range(col, order):
            if abs(mat[row][col]) > abs(mat[pivot_row][col]):
                pivot_row = row

        # æ­¥éª¤2ï¼šè‹¥ä¸»å…ƒä¸º0ï¼Œè¡Œåˆ—å¼å€¼ä¸º0ï¼ˆçŸ©é˜µä¸å¯é€†ï¼‰
        if mat[pivot_row][col] == 0:
            return 0

        # æ­¥éª¤3ï¼šäº¤æ¢å½“å‰è¡Œä¸ä¸»è¡Œï¼ˆè‹¥éœ€è¦ï¼‰ï¼Œå¹¶æ›´æ–°ç¬¦å·
        if pivot_row != col:
            mat[col], mat[pivot_row] = mat[pivot_row], mat[col]
            sign *= -1  # æ¯äº¤æ¢ä¸€æ¬¡è¡Œï¼Œç¬¦å·åè½¬

        # æ­¥éª¤4ï¼šæ¶ˆå»å½“å‰åˆ—ä¸‹æ–¹æ‰€æœ‰å…ƒç´ ï¼ˆåŒ–ä¸ºä¸Šä¸‰è§’çŸ©é˜µï¼‰
        for row in range(col + 1, order):
            factor = mat[row][col] / mat[col][col]  # è®¡ç®—æ¶ˆå…ƒå› å­
            for c in range(col, order):
                mat[row][c] -= factor * mat[col][c]  # è¡Œå˜æ¢ï¼šrow = row - factor*col

    # æ­¥éª¤5ï¼šè®¡ç®—ä¸Šä¸‰è§’çŸ©é˜µå¯¹è§’çº¿å…ƒç´ çš„ä¹˜ç§¯ï¼Œä¹˜ä»¥ç¬¦å·å¾—åˆ°è¡Œåˆ—å¼å€¼
    result = sign
    for i in range(order):
        result *= mat[i][i]

    return round(result)  # å››èˆäº”å…¥å¤„ç†æµ®ç‚¹æ•°ç²¾åº¦è¯¯å·®


# ä¸»ç¨‹åºï¼šè·å–è¾“å…¥å¹¶è®¡ç®—è¡Œåˆ—å¼
if __name__ == "__main__":
    # è¾“å…¥è¡Œåˆ—å¼é˜¶æ•°
    order = int(input("è¯·è¾“å…¥è¡Œåˆ—å¼çš„é˜¶æ•°ï¼š"))
    
    # åˆå§‹åŒ–å¹¶è¾“å…¥è¡Œåˆ—å¼å…ƒç´ 
    det = init_matrix(order)
    for i in range(order):
        for j in range(order):
            det[i][j] = int(input(f"è¯·è¾“å…¥è¡Œåˆ—å¼å…ƒç´  det[{i+1}][{j+1}] = "))
    
    # è®¡ç®—å¹¶è¾“å‡ºç»“æœ
    determinant_value = determinant_calculator(det, order)
    print(f"è¯¥è¡Œåˆ—å¼çš„å€¼ä¸ºï¼š{determinant_value}")
```

```python
def parse_input():
    """
    åŠ¨æ€æ¥æ”¶ç”¨æˆ·è¾“å…¥ï¼Œæ„å»ºè¡Œåˆ—å¼çŸ©é˜µ
    è¾“å…¥è§„åˆ™ï¼šæ¯è¡Œå…ƒç´ ç”¨ç©ºæ ¼åˆ†éš”ï¼ˆå¦‚"1 2 3"ï¼‰ï¼Œè¾“å…¥ç©ºè¡Œæˆ–"end"ç»“æŸè¾“å…¥
    è¿”å›ï¼šæœ‰æ•ˆçš„è¡Œåˆ—å¼çŸ©é˜µï¼ˆæ–¹é˜µï¼‰ï¼Œè‹¥è¾“å…¥æ— æ•ˆåˆ™è¿”å›None
    """
    matrix = []
    print("è¯·é€è¡Œè¾“å…¥è¡Œåˆ—å¼å…ƒç´ ï¼ˆæ¯è¡Œå…ƒç´ ç”¨ç©ºæ ¼åˆ†éš”ï¼Œè¾“å…¥ç©ºè¡Œæˆ–'end'ç»“æŸï¼‰ï¼š")
    
    while True:
        # æ¥æ”¶ä¸€è¡Œè¾“å…¥å¹¶å»é™¤é¦–å°¾ç©ºç™½
        line = input().strip()
        
        # ç»“æŸè¾“å…¥æ¡ä»¶
        if line.lower() == "end" or line == "":
            break
        
        # åˆ†å‰²å…ƒç´ å¹¶è½¬æ¢ä¸ºæ•°å€¼ï¼ˆæ”¯æŒæ•´æ•°å’Œæµ®ç‚¹æ•°ï¼‰
        try:
            # æŒ‰ç©ºæ ¼åˆ†å‰²ï¼Œè¿‡æ»¤ç©ºå­—ç¬¦ä¸²ï¼ˆå¤„ç†è¿ç»­ç©ºæ ¼çš„æƒ…å†µï¼‰
            elements = list(map(float, filter(None, line.split())))
            matrix.append(elements)
        except ValueError:
            print("è¾“å…¥é”™è¯¯ï¼è¯·ç¡®ä¿æ¯è¡Œå…ƒç´ éƒ½æ˜¯æ•°å­—ï¼ˆç”¨ç©ºæ ¼åˆ†éš”ï¼‰ï¼Œé‡æ–°è¾“å…¥è¯¥è¡Œï¼š")
            continue
    
    # æ ¡éªŒæ˜¯å¦ä¸ºæ–¹é˜µï¼ˆè¡Œæ•°=åˆ—æ•°ï¼Œä¸”æ¯è¡Œå…ƒç´ æ•°é‡ç›¸åŒï¼‰
    if not matrix:  # ç©ºçŸ©é˜µ
        print("æœªè¾“å…¥ä»»ä½•å…ƒç´ ï¼Œæ— æ³•è®¡ç®—è¡Œåˆ—å¼ï¼")
        return None
    
    order = len(matrix)  # è¡Œåˆ—å¼é˜¶æ•°=è¡Œæ•°
    for i, row in enumerate(matrix):
        if len(row) != order:
            print(f"è¾“å…¥é”™è¯¯ï¼ç¬¬{i+1}è¡Œæœ‰{len(row)}ä¸ªå…ƒç´ ï¼Œä¸è¡Œæ•°{order}ä¸åŒ¹é…ï¼ˆå¿…é¡»ä¸ºæ–¹é˜µï¼‰ï¼")
            return None
    
    return matrix


def determinant_calculator(matrix):
    """ä½¿ç”¨é«˜æ–¯æ¶ˆå…ƒæ³•è®¡ç®—è¡Œåˆ—å¼å€¼ï¼ˆè¾“å…¥ä¸ºæ–¹é˜µï¼‰"""
    order = len(matrix)
    if order == 0:
        return 0
    
    # æ·±æ‹·è´çŸ©é˜µé¿å…ä¿®æ”¹åŸçŸ©é˜µ
    mat = [row.copy() for row in matrix]
    sign = 1  # è¡Œäº¤æ¢ç¬¦å·ï¼ˆå¶æ•°æ¬¡ä¸º1ï¼Œå¥‡æ•°æ¬¡ä¸º-1ï¼‰
    
    for col in range(order):
        # é€‰ä¸»å…ƒï¼šæ‰¾åˆ°å½“å‰åˆ—ä¸‹æ–¹ï¼ˆå«å½“å‰è¡Œï¼‰ç»å¯¹å€¼æœ€å¤§çš„è¡Œ
        pivot_row = col
        for row in range(col, order):
            if abs(mat[row][col]) > abs(mat[pivot_row][col]):
                pivot_row = row
        
        # ä¸»å…ƒä¸º0åˆ™è¡Œåˆ—å¼å€¼ä¸º0
        if mat[pivot_row][col] == 0:
            return 0
        
        # äº¤æ¢è¡Œå¹¶æ›´æ–°ç¬¦å·
        if pivot_row != col:
            mat[col], mat[pivot_row] = mat[pivot_row], mat[col]
            sign *= -1
        
        # æ¶ˆå»å½“å‰åˆ—ä¸‹æ–¹å…ƒç´ 
        for row in range(col + 1, order):
            factor = mat[row][col] / mat[col][col]
            for c in range(col, order):
                mat[row][c] -= factor * mat[col][c]
    
    # è®¡ç®—å¯¹è§’çº¿ä¹˜ç§¯Ã—ç¬¦å·
    result = sign
    for i in range(order):
        result *= mat[i][i]
    
    return round(result, 6)  # ä¿ç•™6ä½å°æ•°å¤„ç†ç²¾åº¦é—®é¢˜


# ä¸»ç¨‹åº
if __name__ == "__main__":
    # åŠ¨æ€è·å–è¡Œåˆ—å¼çŸ©é˜µ
    det_matrix = parse_input()
    
    # è®¡ç®—å¹¶è¾“å‡ºç»“æœï¼ˆä»…å½“çŸ©é˜µæœ‰æ•ˆæ—¶ï¼‰
    if det_matrix:
        order = len(det_matrix)
        value = determinant_calculator(det_matrix)
        print(f"\n{order}é˜¶è¡Œåˆ—å¼çš„å€¼ä¸ºï¼š{value}")
```
