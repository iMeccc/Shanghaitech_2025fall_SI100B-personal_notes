## 1 CALCULATOR WITH LAMBDA FUNCTIONS (25 POINTS)
### 1.1 Task
Implement a function `calc(op, x, y)`, where `op` is a binary operation function. The function should return the result of `op(x, y)`.

Also, implement a helper function `get_op(op_name)`:  
Given a string `op_name` (such as "add" or "mul"), return the corresponding lambda function. You need to implement the following binary operations using lambda expressions:
- add: addition, \(x+y\)
- sub: subtraction, \(x-y\)
- mul: multiplication, \(x * y\)
- div: division, \(x / y\)
- pow: exponentiation, \(x^{y}\)
- max: return the greater one of the two values
- min: return the smaller one of the two values

### 1.2 Submission
Please submit only your `calc` and `get_op` functions. Do not submit any testing or print statements.

### 1.3 Hint

  ```python

  calc(get_op('add'), 2, 3) == 5

````python
def get_op(op_name):
    match op_name:
        case "add":
            return lambda x,y : x+y
        case "sub":
            return lambda x,y : x-y
        case "mul":
            return lambda x,y : x*y
        case "div":
            return lambda x,y : x/y
        case "pow":
            return lambda x,y : x**y
        case "max":
            return lambda x,y : x if x>=y else y
        case "min":
            return lambda x,y : x if x<=y else y

def calc(op,x,y):
    return op(x,y)
````

## 2 IMAGE LINEAR ENHANCEMENT (25 POINTS)
In 8-bit grayscale imaging, each pixel is an integer in [0, 255]. A common global enhancement is an affine (linear + bias) transform: 
\[y=a \cdot x+b\]

where x is the input pixel, a adjusts contrast (gain), and b shifts brightness (bias). Implementations must also quantize to integers and clip results to a target range so the output stays valid for display or storage.

### 2.1 Task
Given:
- a list of pixel values, each in the range [0, 255];
- real coefficients a and \(b;\)
- an output dynamic range [low, high], which satisfies \(0 ≤ low ≤ high ≤255\)

For each input pixel x , perform the following steps:
Linear transform - compute the adjusted pixel value \(y=a^{*} x+b\)
Quantization — convert y to an integer using the Half-away-from-zero rounding rule:
For \(y ≥0\)

\[y_{q}= \begin{cases}\lfloor y\rfloor, & if y-\lfloor y\rfloor<0.5, \\ \lceil y\rceil, & if y-\lfloor y\rfloor \geq 0.5 .\end{cases}\]

Clipping — ensure the quantized result stays within the valid range [low, high]; any values below low are set to low, and values above high are set to high.
Return the list of enhanced pixels as integers within [low, high].

### 2.2 Input/Output Specification
#### 2.2.1 Input
Four lines:
An integer n represents number of pixels n integers, each in [0, 255] Two reals: a, b Two integers: low ,high (satisfy \(0 ≤ low ≤ high ≤255\) 
\[1 \leq n \leq 100,-1000 \leq a, b \leq 1000 .\]

#### 2.2.2 Output
One line with n integers (the enhanced pixels), separated by single spaces. No trailing spaces.

#### 2.2.3 Example
1. Input
3
10 100 230
1.2 −10 0 255
2. Output
2 110 255


````python
import math

# get n,integers,a,b,low,high
n = int(input())

pixel = list(map(int,input().split(" ")))
a,b = map(float,input().split(' '))
low,high = map(int,input().split(' '))

# do the linear transformation
def linear_trans(pixel,a,b):
    for i in range(len(pixel)):
        pixel[i] = a*pixel[i]+b

# do the quantization
def quant(pixel):
    for i in range(len(pixel)):
        if pixel[i] - math.floor(pixel[i]) < 0.5:
            pixel[i] = math.floor(pixel[i])
        elif pixel[i] - math.floor(pixel[i]):
            pixel[i] = math.ceil(pixel[i])

# do the clipping
def clip(pixel,low,high):
    for i in range(len(pixel)):
        if pixel[i] > high:
            pixel[i] = high
        elif pixel[i] < low:
            pixel[i] = low

# output
def printing(pixel):
    for i in range(len(pixel)-1):
        print(pixel[i],end=' ')
    print(pixel[len(pixel)-1])

linear_trans(pixel,a,b)
quant(pixel)
clip(pixel,low,high)
printing(pixel)
````

## 3 SIMPLIFIED MATCH-3 GAME (25 POINTS)
You are developing a simplified match-3 puzzle game. The game is played on an \(n ×m\) grid where each cell contains a colored block represented by a capital letter. This assignment focuses on implementing basic 2D list operations and simple game mechanics.

### 3.1 Task
Implement a simplified version of a match-3 game with the following rules:
1. Game Board:
• The game is played on an \(n ×m\) grid ( \(3 ≤n\) ,m≤10).
• Each cell contains an uppercase letter (A-Z) representing a colored block.
• The initial board will never contain any pre-existing matches.
2. Game Commands:
• swap x1 y1 x2 y2:
– Swaps the blocks at positions \((x 1, y 1)\) and \((x 2, y 2)\) , where x stands for rows (the input n), and y stands for columns (the input m ).
– Coordinates are 1-indexed (top-left corner is (1, 1)).
• exit:
– Ends the game immediately.
– Outputs final score before terminating.
3. Valid Moves:
• Players can only swap two adjacent blocks (up/down/left/right).
• Diagonal swaps are not allowed.
4. Matching Rules:
• A match occurs when exactly 3 identical blocks line up horizontally or vertically.
• Special Rule: As a simplified match-3 game, the match only counts if the swap places a block in the middle position of the 3-in-a-row, i.e. the swapped block must become the center of 3 identical blocks.
• Each valid match awards exactly 30 points.

5. Match Resolution:
• All 3 matched blocks are removed from the board.
• The empty spaces are filled with ’.’ characters.
• No blocks will fall down to fill empty spaces.
• No chain reactions occur after a match.
6. Invalid Moves:
• Any swap that doesn’t create a valid match (as defined above).
– Attempting to swap non-adjacent blocks.
– Trying to swap blocks outside the grid boundaries.
• All invalid swaps will be automatically undone.

### 3.2 Input/Output Specification
#### 3.2.1 Input
1. First line: Two integers n and m ( \(3 ≤n\) , \(m ≤10)\) ), representing grid dimensions.
2. Next n lines: m space-separated uppercase letters (A-Z) representing initial grid.
3. Subsequent lines: Game commands:
• swap x1 y1 x2 y2 to swap blocks at positions \((x 1, y 1)\) and \((x 2, y 2)\)
• exit to end game.

#### 3.2.2 Output
• To start game, print the current grid and current score.
• After that, for each swap operation print:
1. Current grid state ( n lines of m space-separated characters).
2. Current score.
• After “exit” command print: Final score.

#### 3.2.4 Hint.

````python
def grid_init(n):
    # create a grid
    grid = [[] for _ in range(n)]
    # input the chars
    # input only m chars each line
    for i in range(n):
        grid[i] = input().strip().split(' ')
    return grid

def outprint(grid,score):
    for i in range(len(grid)):
        print(' '.join(grid[i]))
    print(score)

def swap(grid,x1,y1,x2,y2):
    # check validity
    validity = ((y1 == y2 and abs(x1-x2) == 1) or (x1 == x2 and abs(y1-y2) == 1)) and (max(x1,x2) <= len(grid) and max(y1,y2) <= len(grid[0]))
    addscore = 0
    if validity:
        match_1_h = False
        match_1_v = False
        match_2_h = False
        match_2_v = False
        grid[x1-1][y1-1],grid[x2-1][y2-1] = grid[x2-1][y2-1],grid[x1-1][y1-1] # swap the blocks
        # check the first block
        avl_1 = check_idx(x1,y1,grid)
        if avl_1[0]: #水平方向未越界，可以进行比较
            match_1_h = ismatch(grid[x1-1][y1],grid[x1-1][y1-1],grid[x1-1][y1-2])            
        if avl_1[1]: #竖直方向未越界
            match_1_v = ismatch(grid[x1][y1-1],grid[x1-1][y1-1],grid[x1-2][y1-1])      
        # check the second block
        avl_2 = check_idx(x2,y2,grid)
        if avl_2[0]: #水平方向未越界，可以进行比较
            match_2_h = ismatch(grid[x2-1][y2],grid[x2-1][y2-1],grid[x2-1][y2-2])            
        if avl_2[1]: #竖直方向未越界
            match_2_v = ismatch(grid[x2][y2-1],grid[x2-1][y2-1],grid[x2-2][y2-1])

        if match_1_h:
            grid[x1-1][y1],grid[x1-1][y1-1],grid[x1-1][y1-2] = '.','.','.'
            addscore += 3
        if match_1_v:
            grid[x1][y1-1],grid[x1-1][y1-1],grid[x1-2][y1-1] = '.','.','.'
            addscore += 3    
        if match_2_h:
            grid[x2-1][y2],grid[x2-1][y2-1],grid[x2-1][y2-2] = '.','.','.'
            addscore += 3
        if match_2_v:
            grid[x2][y2-1],grid[x2-1][y2-1],grid[x2-2][y2-1] = '.','.','.'
            addscore += 3
        
        # 如果均没有达成合规交换
        if not(match_1_h or match_1_v or match_2_h or match_2_v):
            grid[x1-1][y1-1],grid[x2-1][y2-1] = grid[x2-1][y2-1],grid[x1-1][y1-1] # swap the blocks

    return addscore

def check_idx(x, y, grid):
    horiz = (y >= 2) and (y <= len(grid[0]) - 1)
    vertic = (x >= 2) and (x <= len(grid) - 1)
    return horiz, vertic

def ismatch(nextt,mid,prev):
    return prev == mid == nextt and prev != '.' and mid != '.' and nextt != '.'

grid_size = input().split(' ')
n,m = map(int,grid_size)
grid = grid_init(n)
isexit = False
score = 0
outprint(grid,score)
while not isexit:
    op_line = input().split(' ')
    op_name = op_line[0]
    match op_name:
        case "exit":
            print(score)
            break
        case "swap":
            x1 = int(op_line[1])
            y1 = int(op_line[2])
            x2 = int(op_line[3])
            y2 = int(op_line[4])
            addscore = swap(grid,x1,y1,x2,y2)
            score += 10*addscore
            outprint(grid,score)
````

## 4 F1 RACE STRATEGY ANALYSIS (25 POINTS)
### 4.1 Background
You are an F1 race strategy engineer. During an intense race, you need to analyze telemetry data sent from your driver in real time to make the best strategy decisions such as determining pit-stop windows or evaluating tire performance. You will receive all lap time data of a driver in one race. Each lap contains three sector times (Sector 1, Sector 2, Sector 3) and the pit-stop time at the end of the lap. Your task is to process this raw data and respond to a series of strategy team queries.

### 4.2 Definitions
#### 4.2.1 Lap Time
The pure track time of a lap: 
\[lap_time =S 1+S 2+S 3\]

#### 4.2.2 Total Time
The total time from the end of the previous lap to the end of the current lap: 
\[total_time =S 1+S 2+S 3+ pit_stop_time\]

If there is no pit stop, pit stop time will be 0.0.

#### 4.2.3 Sector Colors
When processing the lap data in sequence, you should keep track of the personal best time (PB) for each of the three sectors (S1, S2, S3). A personal best means the fastest time the driver has achieved in that specific sector so far.
For each new lap:
• If the sector time is smaller (less) than the previous PB, mark this sector as Green. This means the driver has just set a new record for that sector.
• If the sector time is bigger or equal to the previous PB, mark this sector as Yellow. This means the driver did not improve in that sector.

### 4.3 Format
#### 4.3.1 Input
An integer L represents total number of laps at first line. The next L lines: Each lap’s data in the format:
lap_number, s1_time, s2_time, s3_time, pit_stop_time
where:
• lap_number: integer (starting from 0)
• s1_time, s2_time, s3_time: floating numbers
• pit_stop_time: floating number; 0.0 if no pit stop
An integer N represents number of query commands at the next line. The following N lines: Each contains one query instruction (see below).

#### 4.3.2 Output
For each query, print one line of result. All floating values must be rounded to two decimal places.

### 4.4 Query Instructions
| Command | Description | Output Type |
| --- | --- | --- |
| best\_lap | Returns the lap number with the fastest lap time. (If tied, return the smallest lap number.) | Integer |
| pit\_laps | Returns a list of all lap numbers where a pit stop oc- curred. | List |
| lap\_time, i | Returns the lap time of Lap i . | Float |
| best\_sector, s | Returns the all-time personal best time for Sector s . | Float |
| pit\_time, t | Returns a list of lap numbers where a pit stop oc- curred with time ≤ t (ascending order). | List |
| avg\_lap, start, end, step | Returns the average lap time of laps within the slice [start:end:step]. | Float |
| stint\_time, start, end, step | Returns the sum of total time for laps in the slice. | Float |
| green\_laps, s, start, end, step | Returns the list of laps where Sector s was green within the slice (ascending order). | List |
| pit\_rate, start, end, step | Returns the pit-stop rate = (number of pit-stop laps) ÷ (slice length). | Float |
| reorder, key, ASC/DESC | Reorders the current lap dataset by key in ascending ( ASC ) or descending ( DESC ) order. (Ties resolved by lap number ascending.) This command has no out- put . Possible keys: , , s1 , s2 , s3 . | None |

Note: In all multi-parameter queries, start, end, and step refer to index positions. You should first perform slicing on the lap dataset according to these indices ([start,end)). All three values are guaranteed to be integers. If the resulting slice is empty, output: Querying on empty list! and do nothing else.

### 4.6 Hint

````python
class LAP:
    def __init__(self,num,s1,s2,s3,pit):
        self.num = num
        self.s1 = [s1,'']
        self.s2 = [s2,'']
        self.s3 = [s3,'']
        self.pit = pit
        self.lap_time = s1 + s2 + s3
        self.total_time = s1 + s2 + s3 + pit
    

def dataset_init(L):
    dataset = []
    for i in range(L):
        line = input().strip().split(',')
        lap = LAP(int(line[0]),float(line[1]),float(line[2]),float(line[3]),float(line[4]))
        dataset.append(lap)
    return dataset

def best_lap(dataset):
    temp_lap = float('inf')
    best_list = []
    for i in range(len(dataset)):
        if dataset[i].lap_time <= temp_lap:
            temp_lap = dataset[i].lap_time
            best_list.append(dataset[i].num)
    best_list.sort()
    return best_list[0]

def pit_laps(dataset):
    pit_list = []
    for i in range(len(dataset)):
        if dataset[i].pit > 0:
            pit_list.append(dataset[i].num)
    return pit_list

def lap_Time(dataset,i):
    for k in range(len(dataset)):
        if dataset[k].num == i:
            return dataset[k].lap_time

def best_sector(dataset,s):
    s_list = []
    if s == 1:
        for i in range(len(dataset)):
            s_list.append(dataset[i].s1[0])
    if s == 2:
        for i in range(len(dataset)):
            s_list.append(dataset[i].s2[0])
    if s == 3:
        for i in range(len(dataset)):
            s_list.append(dataset[i].s3[0])
    s_list.sort()
    return s_list[0]

def pit_time(dataset,t):
    p_list = []
    for i in range(len(dataset)):
        if dataset[i].pit <= t and dataset[i].pit > 0:
            p_list.append(dataset[i].num)
    return p_list

def avg_lap(dataset,start,end,step):
    index_num = list(range(start,end,step))
    available = False
    avg = 0.0
    if index_num:
        totallap = 0.0
        available = True
        for i in index_num:
            totallap += dataset[i].lap_time
            avg = totallap/len(index_num)
    return available,avg
    
def stint_time(dataset,start,end,step):
    index_num = list(range(start,end,step))
    available = False
    if index_num:
        totallap = 0.0
        available = True
        for i in index_num:
            totallap += dataset[i].lap_time
    return available,totallap
    
def green_laps(dataset,s,start,end,step):
    index_num = list(range(start,end,step))
    available = False
    if index_num:
        green_list = []
        available = True
        if s == 1:
            for i in index_num:
                if dataset[i].s1[1] == "Green":
                    green_list.append(dataset[i].num)
        if s == 2:
            for i in index_num:
                if dataset[i].s2[1] == "Green":
                    green_list.append(dataset[i].num)
        if s == 3:
            for i in index_num:
                if dataset[i].s3[1] == "Green":
                    green_list.append(dataset[i].num)
        green_list.sort()
    return available,green_list
        
def pit_rate(dataset,start,end,step):
    index_num = list(range(start,end,step))
    available = False
    rate = 0.0
    if index_num:
        pit_stop = 0
        available = True
        for i in index_num:
            if dataset[i].pit > 0:
                pit_stop += 1
        rate = pit_stop/len(index_num)
    return available,rate
    
def reorder(dataset, keys, op):
    def get_sort_key(x):
        attr = getattr(x, keys)
        if keys in ['s1', 's2', 's3']:
            return attr[0]
        return attr
    dataset.sort(key=lambda x: (get_sort_key(x), x.num),reverse=(op == "DESC"))
    new_num = 0
    for lap in dataset:
        lap.num = new_num
        new_num += 1

def name_color(dataset):
    temp_pb_1 = float('inf')
    temp_pb_2 = float('inf')
    temp_pb_3 = float('inf')
    for i in range(len(dataset)):
        if dataset[i].s1[0] < temp_pb_1:
            dataset[i].s1[1] = "Green"
            temp_pb_1 = dataset[i].s1[0]
        else:
            dataset[i].s1[1] = "Yellow"
        if dataset[i].s2[0] < temp_pb_2:
            dataset[i].s2[1] = "Green"
            temp_pb_2 = dataset[i].s2[0]
        else:
            dataset[i].s2[1] = "Yellow"
        if dataset[i].s3[0] < temp_pb_3:
            dataset[i].s3[1] = "Green"
            temp_pb_3 = dataset[i].s3[0]
        else:
            dataset[i].s3[1] = "Yellow"

L = int(input())
dataset = dataset_init(L)
name_color(dataset)

N = int(input())
for i in range(N):
    line = input().strip().split(',')
    instruction = line[0]
    match instruction:
        case "best_lap":
            print(best_lap(dataset))
        case "pit_laps":
            print(pit_laps(dataset))
        case "lap_time":
            i = int(line[1])
            print(f"{lap_Time(dataset,i):.2f}")
        case "best_sector":
            s = int(line[1])
            print(f"{best_sector(dataset,s):.2f}")
        case "pit_time":
            t = float(line[1])
            print(pit_time(dataset,t))
        case "avg_lap":
            start = int(line[1])
            end = int(line[2])
            step = int(line[3])
            available = avg_lap(dataset,start,end,step)
            if available[0]:
                print(f"{available[1]:.2f}")
            else:
                print("Querying on empty list!")
        case "stint_time":
            start = int(line[1])
            end = int(line[2])
            step = int(line[3])
            available = stint_time(dataset,start,end,step)
            if available[0]:
                print(f"{available[1]:.2f}")
            else:
                print("Querying on empty list!")
        case "green_laps":
            s = int(line[1])
            start = int(line[2])
            end = int(line[3])
            step = int(line[4])
            available = green_laps(dataset,s,start,end,step)
            if available[0]:
                print(available[1])
            else:
                print("Querying on empty list!")
        case "pit_rate":
            start = int(line[1])
            end = int(line[2])
            step = int(line[3])
            available = pit_rate(dataset,start,end,step)
            if available[0]:
                print(f"{available[1]:.2f}")
            else:
                print("Querying on empty list!")
        case "reorder":
            keys = str(line[1])
            op = str(line[2])
            reorder(dataset,keys,op)
````
