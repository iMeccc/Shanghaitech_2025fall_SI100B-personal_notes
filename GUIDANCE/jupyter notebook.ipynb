## 1. Notebook 基本结构
Notebook 由一系列单元（cells）组成，主要有两类：
- Markdown 单元：用于书写文本、数学公式、图片与说明，适合做讲义要点与复习要点；
- Code 单元：运行代码（例如 Python），可直接在单元中执行并查看输出；

每个单元都可以单独运行，运行结果保存在单元下方，这使得分块记录实验、示例与解释非常方便。

## 2. 在 VS Code 中使用 Notebook
- 打开 `.ipynb` 文件：直接在 VS Code 资源管理器中打开；
- 运行单元：在每个单元左上角有 `Run Cell` 按钮，或使用快捷键；
- 在 `.py` 文件中使用交互单元：可以在 `.py` 文件中插入 `# %%` 来标记单元并在 Interactive 窗口中运行；

推荐插件：Python（Microsoft），Jupyter（Microsoft），Markdown All in One。

**课堂问题 & 任务**：

**复习要点**：
- 考试可能覆盖的要点


````python
# 简单 Python 示例：计算经验熵（示意）
import random, math
def empirical_entropy(trials=1000):
    counts = {0:0,1:0}
    for _ in range(trials):
        r = random.random() < 0.5
        counts[int(r)] += 1
    p0 = counts[0]/trials
    p1 = counts[1]/trials
    return -sum(p*math.log2(p) for p in (p0,p1) if p>0)

print(empirical_entropy(10000))
````

## 4. 常用快捷键与导出
- 运行当前单元（Run Cell）：点击单元左侧的运行按钮或使用 UI 中的命令；
- 运行全部单元（Run All）：使用命令面板中的 Run All；
- 在 VS Code 中预览 Markdown：Ctrl+Shift+V；
- 导出为 PDF：在预览窗口中打印为 PDF，或使用 Pandoc 导出（需安装 Pandoc 与 LaTeX）。
